# 모듈 불러오기

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
mglearn.plots.plot_tree_progressive()
```

# 데이터 적재

```python
cancer = load_breast_cancer()
```

## 살펴보기

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)
```

## 학습 및 모델


```python
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)
```

## 평가 및 검증

```python
tree.score(X_train, y_train), tree.score(X_test, y_test)
```

```python
plt.figure(figsize=(12,10))
plot_tree(tree, class_names=['악성', '양성'], feature_names=cancer.feature_names.tolist(), impurity=False, filled=True, rounded=True, fontsize=10)
```

```python
def plot_feature_importances_cancer(model):
    n_features = cancer.data.shape[1]
    plt.barh(np.arange(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), cancer.feature_names)
    plt.xlabel("특성 중요도")
    plt.ylabel("특성")
    plt.ylim(-1, n_features)

plt.figure(figsize=(12,10))
plot_feature_importances_cancer(tree)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=100, noise=0.25, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
forest = RandomForestClassifier(n_estimators=5, random_state=42)
forest.fit(X_train, y_train)
```

```python
fig, axes = plt.subplots(2,3,figsize=(20,20))
for i, (ax, _tree) in enumerate(zip(axes.ravel(), forest.estimators_)):
    ax.set_title(f"트리 {i}")
    mglearn.plots.plot_tree_partition(X, y, _tree, ax=ax)

mglearn.plots.plot_2d_separator(forest, X, fill=True, ax=axes[-1, -1], alpha=.4)
axes[-1, -1].set_title("랜덤 포레스트")
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
```
